```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_breast_cancer
```

```python
l=load_breast_cancer()
```

```python
df=pd.DataFrame(data=load_breast_cancer().data,columns=load_breast_cancer().feature_names)
```

```python
df.columns
```

```python
df["label"]=l.target
```

```python
f,ax=plt.subplots(figsize=(18,18))
sns.heatmap(df.corr(),annot=True,linewidths=0.5,linecolor="red",fmt="0.1f",ax=ax)
```

```python
y=df["label"].values
x=df.iloc[:,:-1].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
decision_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
svc = SVC()
NBbyes = GaussianNB()
```

```python
decision_tree.fit(x_train,y_train)
random_forest.fit(x_train,y_train)
svc.fit(x_train,y_train)
NBbyes.fit(x_train,y_train)
```

```python
pred1 = decision_tree.predict(x_test)
pred2 = random_forest.predict(x_test)
pred3 = svc.predict(x_test)
pred4 = NBbyes.predict(x_test)
```

```python
print("The Accuracy of the Decision Tree is {}".format(accuracy_score(y_test,pred1)))
print("The Accuracy of the Random Forest is {}".format(accuracy_score(y_test,pred2)))
print("The Accuracy of the SVC is {}".format(accuracy_score(y_test,pred3)))
print("The Accuracy of the NBbyes is {}".format(accuracy_score(y_test,pred4)))
```
